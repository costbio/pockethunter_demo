```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('docking4_results.csv', delimiter=',', decimal='.')
```

```python
df
```

```python
# Remove rows with missing values
df.dropna(inplace=True)

# Assign inhibitor labels based on ligand name
df['Inhibitor'] = df['Ligand'].str.startswith('active')

# Calculate mean docking score across all targets for each ligand
score_columns = [col for col in df.columns if col not in ['Ligand', 'Inhibitor']]
df['mean_score'] = df[score_columns].mean(axis=1)

# Sort by mean docking score (lower scores = better binding)
df.sort_values('mean_score', inplace=True)

df
```

```python
df
```

```python
# Compare ROC curves for each protein and mean_score
from sklearn.metrics import roc_curve, auc
import numpy as np

plt.figure(figsize=(8,6))
score_columns = [col for col in df.columns if col not in ['Ligand', 'Inhibitor', 'mean_score']]

for col in score_columns:
    y_true = df['Inhibitor'].astype(int)
    y_score = -df[col]  # Negate so higher score = more likely active
    fpr, tpr, _ = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{col} (AUC = {roc_auc:.2f})')

# Mean score ROC
y_score_mean = -df['mean_score']
fpr_mean, tpr_mean, _ = roc_curve(y_true, y_score_mean)
roc_auc_mean = auc(fpr_mean, tpr_mean)
plt.plot(fpr_mean, tpr_mean, label=f'Mean Score (AUC = {roc_auc_mean:.2f})', linewidth=2, color='black')

plt.plot([0, 1], [0, 1], 'k--', label='Random baseline')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Docking Scores (per protein and mean)')
plt.legend(loc='lower right')


# Try moving the text to a different location to see if it appears
plt.text(-0.04, 1, 'Exhaustiveness = 32', fontsize=10, color='black')
plt.show()
```

```python
# Best-practice ROC analysis: summary table, ROC curves, score distributions
from sklearn.metrics import roc_curve, auc
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Your existing code to set up data and calculate AUCs
# Assuming df is already defined
score_columns = [col for col in df.columns if col not in ['Ligand', 'Inhibitor', 'mean_score']]
y_true = df['Inhibitor'].astype(int)

# 1. Calculate AUCs for each protein and mean score
auc_dict = {}
for col in score_columns:
    y_score = -df[col]
    fpr, tpr, _ = roc_curve(y_true, y_score)
    auc_dict[col] = auc(fpr, tpr)
y_score_mean = -df['mean_score']
fpr_mean, tpr_mean, _ = roc_curve(y_true, y_score_mean)
auc_dict['mean_score'] = auc(fpr_mean, tpr_mean)

auc_df = pd.DataFrame(list(auc_dict.items()), columns=['Score', 'AUC'])
print('AUC summary table:')
display(auc_df.sort_values('AUC', ascending=False))

# --- Modified plotting code starts here ---

# 3. Plot score distributions for actives vs. inactives
# Create subplots in a 3x3 grid
fig, axes = plt.subplots(3, 3, figsize=(16,12)) # Removed the outer plt.figure() as subplots creates it
# Flatten the 2D array of axes into a 1D array for easy iteration
axes = axes.flatten()

# Add a super title to the entire figure
fig.suptitle('Score Distributions for Actives vs. Inactives (Exhaustiveness = 64)', fontsize=16, y=1.02) # y controls vertical position

num_plots = len(score_columns) + 1  # +1 for mean_score
for i, col in enumerate(score_columns + ['mean_score']):
    ax = axes[i] # Get the specific axes object for this plot
    
    sns.histplot(data=df[df['Inhibitor'] == True], x=col, ax=ax, color='blue', label='Active', kde=True, stat='density', bins=20)
    sns.histplot(data=df[df['Inhibitor'] == False], x=col, ax=ax, color='red', label='Inactive', kde=True, stat='density', bins=20)
    
    ax.set_title(f'{col}') # Title for each subplot, might want to simplify as suptitle handles the main context
    ax.set_xlabel('Docking Score')
    ax.legend()

# Remove any unused subplots
for i in range(num_plots, len(axes)):
    axes[i].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.98]) # Adjust tight_layout to make space for the suptitle
plt.show()
```
