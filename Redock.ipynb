```python
import requests
import xml.etree.ElementTree as ET
from Bio.PDB import PDBList
import subprocess
import os

def get_pdb_ids_from_uniprot(uniprot_id):
    url = f"https://www.uniprot.org/uniprot/{uniprot_id}.xml"
    response = requests.get(url)
    tree = ET.ElementTree(ET.fromstring(response.content))
    root = tree.getroot()
    
    pdb_ids = []
    for cross_ref in root.findall(".//{http://uniprot.org/uniprot}dbReference[@type='PDB']"):
        pdb_ids.append(cross_ref.attrib['id'])
    
    return pdb_ids


uniprot_id = 'P53350'  # Uniprot ID FOR VEGFR2

# ACQUIRING THE PDB ID'S
pdb_ids = get_pdb_ids_from_uniprot(uniprot_id)
```

```python
import csv

pocket_preferences = {
    "4A4L": 2,
    "2YAC": 2,
    "4J52": 2,
    "3FC2": 2,
    "2rku_A_prod_R3_fit_4180": 1,
    "2rku_A_prod_R3_fit_1752": 2, #added later on
    "2rku_A_prod_R3_fit_3147": 1, #added later on
    "2RKU": 1
}


atomiddict = {}
for pdb in pdb_ids:
    #all ligands under the ligands_pdbqt_dir

    for lig in os.listdir(ligands_pdbqt_dir):
        ligf = lig.split(".")[0] 
        print(pdb, ligf)

        #create a directory for docking simulations
        docking_dir = f"{current_directory}/docking4/{pdb}/{ligf}"
        os.makedirs(docking_dir, exist_ok=True)
        #copy the pdbqt files of protein and ligand to the docking directory
        ##shutil.copy(f"{current_directory}/pdbqt_list/{pdb}.pdbqt", docking_dir)
        ##shutil.copy(f"{ligands_pdbqt_dir}/{lig}", docking_dir)

        #lets take the p2rank data for pocket center and coordinates
        output_dir = f'{current_directory}/detect/p2rank_output'
        predictions_file = os.path.join(output_dir, os.path.basename(pdb+".pdb_predictions.csv"))

        target_pocket_rank = pocket_preferences.get(pdb, 1)

        #read the csv file and get the coordinates of the top pocket
        df = pd.read_csv(predictions_file)
        # Stun adlarndaki boluklar kaldrma
        df.columns = df.columns.str.strip()

        top_pocket = df.iloc[target_pocket_rank - 1]
        center_x = top_pocket['center_x']
        center_y = top_pocket['center_y']
        center_z = top_pocket['center_z']

        
        with open(predictions_file, 'r') as pred_file:
            reader = csv.DictReader(pred_file)
            for row in reader:
                if row['name     '].strip() == f'pocket{target_pocket_rank}':
                    atomids = row[" surf_atom_ids"]
                    atomids = atomids[1:]
                    atomids = atomids.replace(" ",",")
            #atomids = row[" residue_ids"].strip().split()
                    # Remove any prefix like "A_", "B_", "C_", etc.
                    #atomids = [re.sub(r'^[A-Z]_', '', atomid) for atomid in atomids]
                    #print(",".join(atomids))

                    atomiddict[pdb] = atomids
                    print(atomids)

        #now create a config file for each docking simulation
        with open(f"{docking_dir}/config.txt", "w") as f:
            f.write(f"receptor = {docking_dir}/{pdb}.pdbqt\n")
            f.write(f"ligand = {docking_dir}/{ligf}.pdbqt\n")
            f.write("exhaustiveness = 64\n")
            f.write("num_modes = 9\n")
            f.write("energy_range = 3\n")
            f.write(f"center_x = {center_x}\n")
            f.write(f"center_y = {center_y}\n")
            f.write(f"center_z = {center_z}\n")
            f.write(f"out = {docking_dir}/{pdb}_{ligf}_out.pdbqt\n")
        




        
```

```python
from Bio.PDB import PDBParser
saved_coordinates={}
for pdbname, surf_atom_ids in atomiddict.items():

    #surf_atom_ids was str when imported, so i turned into int to make this code work
    surf_atom_ids = list(map(int, surf_atom_ids.split(',')))

    
    pdb_file = f"{current_directory}/pdb_list/{pdbname}.pdb"
    parser = PDBParser()
    structure = parser.get_structure("protein", pdb_file)

    # Initialize an empty dictionary to store atom coordinates
    atom_coordinates = {}
    # Iterate through all atoms and extract coordinates
    for model in structure:
        for chain in model:
            for residue in chain:
                for atom in residue:
                    if atom.get_serial_number() in surf_atom_ids:
                        atom_coordinates[atom.get_serial_number()] = atom.get_coord()
                        saved_coordinates[pdbname] = atom_coordinates

    print(f" {pdbname} of the coordinates of selected atoms:", atom_coordinates)

    print()
```

```python
import numpy as np

# Example dictionary of atom coordinates
def gridbox_calculate(atom_coordinates):
    # Extract all x, y, z coordinates into separate arrays
    x_coords = np.array([coord[0] for coord in atom_coordinates.values()])
    y_coords = np.array([coord[1] for coord in atom_coordinates.values()])
    z_coords = np.array([coord[2] for coord in atom_coordinates.values()])

    # Find the min and max for each axis
    x_min, x_max = np.min(x_coords), np.max(x_coords)
    y_min, y_max = np.min(y_coords), np.max(y_coords)
    z_min, z_max = np.min(z_coords), np.max(z_coords)

    # Calculate the center of the grid box
    var_center_x = (x_min + x_max) / 2
    var_center_y = (y_min + y_max) / 2
    var_center_z = (z_min + z_max) / 2

    # Calculate the size of the grid box
    var_size_x = x_max - x_min
    var_size_y = y_max - y_min
    var_size_z = z_max - z_min

    # Print the results
    centers = f"center_x = {var_center_x}\ncenter_y = {var_center_y}\ncenter_z = {var_center_z}\n"
    gridbox = f"size_x = {var_size_x}\nsize_y = {var_size_y}\nsize_z = {var_size_z}\n"
    #print(f"Size of the grid box: ({var_size_x}, {var_size_y}, {var_size_z})")
    return centers,gridbox
```
